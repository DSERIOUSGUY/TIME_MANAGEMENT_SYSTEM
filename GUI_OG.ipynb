```python
parent=Tk()


label1=Label(parent,text="Emp ID")
label1.grid(row=2,column=1)

label2=Label(parent,text="Password")
label2.grid(row=3,column=1)

TextField1=Entry(parent)
TextField1.grid(row=2,column=4)

TextField2=Entry(parent)
TextField2.grid(row=3,column=4)

label3=Label(parent,text="")
label3.grid(row = 6, column= 2)
   
    ####################################
    ##LOGIN##
    
def login(lab,entry1,entry2):
    res1=entry1.get()
    res2=entry2.get()
    cr=db.cursor()
    cr.execute('select password from Employee where Emp_ID = %d' % int(res1))
    for i in cr:
        if(i[0]==res2):
            mainBox()
        else:
            lab.configure(text='Incorrect Emp ID / password')
    
    
    
button1=Button(parent,text="login", command = lambda: login(label3,TextField1,TextField2))
button1.grid(row=8,column=1)    
       
        
    
    #####################################  
    ##SIGNUP##
    
def newBox():
    
    box = Toplevel(parent)
    lab1=Label(box,text="Name")
    lab1.grid(row=2,column=1)
    lab2=Label(box,text="Password")
    lab2.grid(row=3,column=1)
    lab3=Label(box,text="email")
    lab3.grid(row=4,column=1)
    lab4=Label(box,text="")
    lab4.grid(row=5,column=3)
    Text1=Entry(box)
    Text1.grid(row=2,column=4)
    Text2=Entry(box)
    Text2.grid(row=3,column=4)
    Text3=Entry(box)
    Text3.grid(row=4,column=4)
    but=Button(box,text="confirm", command= lambda:readInp(box,lab4,Text1,Text2,Text3))
    but.grid(row=8,column=3)
    box.mainloop()
    
def readInp(box,lab4,entry1,entry2,entry3):
    res1=entry1.get()
    res2=entry2.get()
    res3=entry3.get()
    cr=db.cursor()
    cr.execute('insert into Employee(_name,password,email) values(%s,%s,%s);',(res1,res2,res3))
    lab4.configure(text='Registerd. Your Emp ID is: ' + str(cr.execute('select Emp_ID from Employee order by Emp_ID;')+2))
    db.commit()
    
    
button2=Button(parent,text="signup", command= newBox)
button2.grid(row=8,column=4)

    #######################################
    ##MAIN##

def mainBox():
    parent.withdraw()
    mainBox = Toplevel(parent)
     
    mainBut1=Button(mainBox,text='Check Schedule',command=lambda:checkSchedule())
    mainBut1.grid(row=1,column=1)
    
    #mainBut2=Button(mainBox,text='Schedule Meeting',command=lambda:scheduleMeeting())
    #mainBut2.grid(row=1,column=2)
    
    mainBut3=Button(mainBox,text='Cancel Meeting',command=lambda:cancelMeeting())
    mainBut3.grid(row=1,column=2)
    
    mainBut4=Button(mainBox,text='Logout',command=lambda: logout(mainBox))
    mainBut4.grid(row=1,column=3)
    
    lab = Label(mainBox, text = "* Always logout before quitting! *")
    lab.grid(row=3, column=2)
    
    mainBox.mainloop()
    
def logout(box):
    box.destroy()
    parent.deiconify()
    
def checkSchedule():
    schedule=Toplevel(parent)
    schLab1=Label(schedule,text="")
    schLab1.grid(row=1,column=0)
    cr=db.cursor()
    cr.execute('delete from timeTable where _date < CURDATE();')
    schLab1.configure(text='You have '+str(cr.execute('select * from timeTable where _date = CURDATE();'))+' meetings today' )
    #scrollbar = tk.Scrollbar(schedule,orient='vertical')
    Lb = Listbox(schedule, width=50, height=20)
    x=0
    for i in cr:
        x=x+1
        Lb.insert(x,i)
    Lb.grid()
    scBut1=Button(schedule,text="view all meetings",command = lambda: allMeetings(Lb))
    scBut1.grid(row=3, column=0)
    scBut2=Button(schedule,text="view today's meetings",command = lambda: todayMeetings(Lb))
    scBut2.grid(row=5, column=0)
    mainBut2=Button(schedule,text='Schedule Meeting',command=lambda:scheduleMeeting(Lb))
    mainBut2.grid(row=4,column=0)
    schedule.mainloop()
    
def allMeetings(Lb):
    Lb.delete(0,'end')
    cr.execute('select * from timeTable;')
    x=0
    for i in cr:
        x=x+1
        Lb.insert(x,i)
    
    
    
    
def todayMeetings(Lb):
    Lb.delete(0,'end')
    cr.execute('select * from timeTable where _date = CURDATE();')
    x=0
    for i in cr:
        x=x+1
        Lb.insert(x,i)
   


# def removeMeeting(Lb):
#     Lb.delete(0,'end')
#     y=Lb.get(tk.ACTIVE)
#     x=0
#     for i in cr:
#         x=x+1
#         if (y==cr):
#             while (x>0):
#                 cr=cr+1
#             cr.execute('')


def scheduleMeeting(Lb):
    schMeeting=Tk()
    txtField1=Entry(schMeeting)
    txtField1.grid(row = 2, column = 2)
    txtField2=Entry(schMeeting)
    txtField2.grid(row = 4, column = 2)
    txtField3=Entry(schMeeting)
    txtField3.grid(row = 6, column = 2)
    txtField4=Entry(schMeeting)
    txtField4.grid(row = 8, column = 2)
    
    lab1= Label(schMeeting, text = 'Date')
    lab1.grid(row = 2, column =1)
    lab2= Label(schMeeting, text = 'from')
    lab2.grid(row = 4, column =1)
    lab3= Label(schMeeting, text = 'to')
    lab3.grid(row = 6, column =1)
    lab4= Label(schMeeting, text = 'Client')
    lab4.grid(row = 8, column =1)
    but1= Button(schMeeting, text = 'make appointment',command = lambda:makeAppointment(schMeeting,txtField1,txtField2,txtField3,txtField4))
    but1.grid(row= 12, column = 1)
    schMeeting.mainloop()
    
def makeAppointment(box,entry1,entry2,entry3,entry4):
    res1=datetime.strptime(entry1.get(),'%y-%m-%d').date()
    res2=datetime.strptime(entry2.get(),'%H:%M:%S').time()
    res3=datetime.strptime(entry3.get(),'%H:%M:%S').time()
    res4=entry4.get()
    cr.execute('insert into timeTable(_date,_from,_to) values(%s,%s,%s) ;', (res1,res2,res3)) ## <<I add client in sql and code here
    x=Tk()
    lab = Label(x, text='Appointment Scheduled')
    lab.grid(row = 1 , column = 2)
    butConf=Button(x,text = 'ok', command = lambda :ok(x,box))
    butConf.grid(row=2 , column =2)
    x.mainloop()
def ok(x,y):
    x.destory()
    y.destroy()
    
##AttributeError: '_tkinter.tkapp' object has no attribute 'destory'        
    
    
        
        


    
    
    
    
    
    
    


    
parent.mainloop()
```
